## Setup — Run this cell first


```python
import pandas as pd
import numpy as np

# World Happiness Report 2023 — representative data
# Source: https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2023

df = pd.read_csv('../data/world_happiness_2023.csv')
df.columns = ['Country','Region','Happiness_Score','GDP','Social_Support',
              'Life_Expectancy','Freedom','Generosity','Corruption']


print(f"Dataset: {len(df)} countries, {len(df.columns)} columns")
print(df.head())
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Explore the dataset before you start
print("Regions in dataset:")
print(df['Region'].value_counts())
print("\nScore range:", df['Happiness_Score'].min(), "–", df['Happiness_Score'].max())
print("\nBottom 10 countries:")
print(df.nsmallest(10, 'Happiness_Score')[['Country','Region','Happiness_Score']])
```

## Task 1 — Regional Comparison Bar Chart

**What to build:** A horizontal bar chart showing the **average happiness score by region**, sorted from highest to lowest.

**Requirements:**
- Horizontal orientation (category names are long)
- Sorted by score, descending (so the happiest region is at the top)
- Zero baseline on x-axis
- At least one design choice that goes beyond the Plotly default (colour, annotation, labels, etc.)
- An insight title that answers: *which region stands out and why does it matter?*

**Hint:** Use `df.groupby('Region')['Happiness_Score'].mean()` to compute the averages.


```python
# Task 1: Regional comparison bar chart
# -------------------------------------

# Step 1: Compute average happiness score by region
region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score'))  # sort for horizontal bar

print(region_avg)

# Step 2: Build your chart
# YOUR CODE HERE

region_avg = (df.groupby('Region')['Happiness_Score']
              .mean()
              .reset_index()
              .sort_values('Happiness_Score', ascending=True))  # ascending=True for horizontal bar (top = highest)

colors = ['#F0997B' if s < 5.5 else '#5DCAA5' if s < 6.5 else '#0F6E56'
          for s in region_avg['Happiness_Score']]

fig = go.Figure(go.Bar(
    x=region_avg['Happiness_Score'],
    y=region_avg['Region'],
    orientation='h',
    marker_color=colors,
    text=region_avg['Happiness_Score'].round(2),
    textposition='outside',
))

fig.update_layout(
    title=dict(text="North America & Western Europe lead global happiness by a wide margin",
               font=dict(size=15, color='#2C2C2A')),
    xaxis=dict(
        range=[0, 8.5],
        title=dict(text="Average Happiness Score", font=dict(size=13, color='#2C2C2A')),
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='#444441',
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=True,
        gridcolor='#D3D1C7',
        gridwidth=1,
    ),
    yaxis=dict(
        title="",
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=False,
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=420,
    margin=dict(l=210, r=90, t=60, b=50),
    font=dict(family="Arial", size=12),
)
fig.show()

```

## Task 2 — Bottom vs. Top: A Contrast Story

**What to build:** A bar chart that highlights the **gap between the happiest and least happy countries**, focusing on a specific insight.

**Requirements:**
- Show the **top 8 AND bottom 8 countries** together (16 bars total)
- Use **colour** to distinguish the two groups (not Plotly's default rainbow)
- Add a **visual separator or annotation** that emphasises the gap
- Insight title that tells the story of the gap

**Hint:** Use `pd.concat([df.nlargest(8,'Happiness_Score'), df.nsmallest(8,'Happiness_Score')])` to get both groups.

**Stretch goal:** Add a vertical reference line showing the global average.


```python
# Task 2: Top 8 vs. Bottom 8 contrast
# ------------------------------------

# Step 1: Get top and bottom countries
top8 = df.nlargest(8, 'Happiness_Score').copy()
top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy()
bottom8['Group'] = 'Bottom 8'

combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score')
global_avg = df['Happiness_Score'].mean()
print(f"Global average: {global_avg:.2f}")

# Step 2: Build your chart
# YOUR CODE HERE

top8 = df.nlargest(8, 'Happiness_Score').copy(); top8['Group'] = 'Top 8'
bottom8 = df.nsmallest(8, 'Happiness_Score').copy(); bottom8['Group'] = 'Bottom 8'
combined = pd.concat([bottom8, top8]).sort_values('Happiness_Score', ascending=True)  # ascending for horiz
global_avg = df['Happiness_Score'].mean()

color_map = {'Top 8': '#0F6E56', 'Bottom 8': '#993C1D'}

fig = px.bar(combined, x='Happiness_Score', y='Country',
             orientation='h', color='Group',
             color_discrete_map=color_map,
             text='Happiness_Score',
             category_orders={'Country': combined['Country'].tolist()})

fig.update_traces(texttemplate='%{text:.3f}', textposition='outside')

fig.add_vline(x=global_avg, line_dash='dash', line_color='#BA7517', line_width=2,
              annotation_text=f"Global avg: {global_avg:.2f}",
              annotation_position="top right",
              annotation_font=dict(color='#BA7517', size=12))

fig.update_layout(
    title=dict(text="The world's happiest and least happy nations are separated by nearly 6 points",
               font=dict(size=15, color='#2C2C2A')),
    xaxis=dict(
        range=[0, 9.5],
        title=dict(text="Happiness Score", font=dict(size=13, color='#2C2C2A')),
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='#444441',
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=True,
        gridcolor='#D3D1C7',
        gridwidth=1,
    ),
    yaxis=dict(
        title="",
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=False,
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=640,
    margin=dict(l=150, r=110, t=60, b=50),
    legend=dict(title="", orientation='h', y=-0.07, font=dict(size=12)),
    font=dict(family="Arial", size=12),
)
fig.show()

```

## Done? Stretch Goal

If you finish both tasks with time to spare, try this:

**Task 3 (stretch):** Build a **grouped bar chart** comparing 2 sub-factors (e.g. `GDP_per_capita` and `Freedom`) across the 5 most populated regions. Use colour meaningfully and write an insight title.

Regions to include: `'Western Europe'`, `'Latin America'`, `'East Asia'`, `'Sub-Saharan Africa'`, `'South Asia'`


```python
# Stretch goal — grouped bar chart
# YOUR CODE HERE

selected_regions = ['Western Europe', 'Latin America and Caribbean',
                    'East Asia', 'Sub-Saharan Africa', 'South Asia']
stretch_df = df[df['Region'].isin(selected_regions)]
grouped = (stretch_df.groupby('Region')[['GDP', 'Freedom']]
           .mean()
           .reset_index()
           .sort_values('GDP', ascending=True))  # ascending=True so highest appears at top

fig = go.Figure()
fig.add_trace(go.Bar(name='GDP per capita (log)', y=grouped['Region'],
                     x=grouped['GDP'], orientation='h',
                     marker_color='#185FA5',
                     text=grouped['GDP'].round(2), textposition='outside'))
fig.add_trace(go.Bar(name='Freedom score', y=grouped['Region'],
                     x=grouped['Freedom'], orientation='h',
                     marker_color='#5DCAA5',
                     text=grouped['Freedom'].round(2), textposition='outside'))

fig.update_layout(
    barmode='group',
    title=dict(text="Wealthy regions lead on GDP; freedom scores are more evenly distributed",
               font=dict(size=15, color='#2C2C2A')),
    xaxis=dict(
        range=[0, 13],
        title=dict(text="Score", font=dict(size=13, color='#2C2C2A')),
        zeroline=True,
        zerolinewidth=2,
        zerolinecolor='#444441',
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=True,
        gridcolor='#D3D1C7',
        gridwidth=1,
    ),
    yaxis=dict(
        title="",
        linecolor='#444441',
        linewidth=2,
        tickfont=dict(size=12, color='#2C2C2A'),
        showgrid=False,
    ),
    plot_bgcolor='white',
    paper_bgcolor='white',
    height=420,
    margin=dict(l=200, r=90, t=60, b=50),
    legend=dict(orientation='h', y=-0.12, font=dict(size=12)),
    font=dict(family="Arial", size=12),
)
fig.show()
```
